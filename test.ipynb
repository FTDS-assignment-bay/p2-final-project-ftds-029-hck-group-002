```python
import pandas as pd

# Load dataset
df1 = pd.read_csv('AI_Resume_Screening.csv')
df2 = pd.read_csv('ResumeDataSet.csv')

# --- 1️⃣ Samakan nama kolom yang berbeda tapi artinya sama ---
# Misalnya kolom di df2 "Resume" artinya sama dengan "Resume Text"
rename_map = {
    'Resume': 'Resume Text',
    'Category': 'Job Category'
}
df2.rename(columns=rename_map, inplace=True)

# --- 2️⃣ Gabungkan kolom (outer join on all columns) ---
df_combined = pd.concat([df1, df2], ignore_index=True, sort=False)

# --- 3️⃣ Simpan ke file baru ---
df_combined.to_csv('combined_resume_dataset.csv', index=False)

# --- 4️⃣ Lihat hasil ---
print("Shape:", df_combined.shape)
df_combined.head()
```

```python
# Load the datasets
df1 = pd.read_csv('AI_Resume_Screening.csv')
df2 = pd.read_csv('ResumeDataSet.csv')

# Cek nama kolom di masing-masing dataset
cols1 = set(df1.columns)
cols2 = set(df2.columns)

# Mapping kolom yang mirip (berdasarkan nama & konteks)
rename_map = {}
if 'Resume' in df2.columns:
    rename_map['Resume'] = 'Resume Text'
if 'Category' in df2.columns:
    rename_map['Category'] = 'Job Category'

# Rename kolom df2
df2.rename(columns=rename_map, inplace=True)

# Gabungkan semua kolom (outer join)
df_combined = pd.concat([df1, df2], ignore_index=True, sort=False)

# Simpan hasil
output_path_aligned = 'combined_resume_dataset_aligned.csv'
df_combined.to_csv(output_path_aligned, index=False)

```
